```python
#importing modules

import pandas as pd
from matplotlib import pyplot as plt
```

```python
#load file and choosing specific columns only

df=pd.read_csv("train.csv", usecols=["PassengerId", "Survived", "Pclass", "Name", "Sex", "Age"])
df.tail(10)
df.describe()
```

```python
#sorting the values by Age

df=df.sort_values("Age", ascending=False)
df.head(10)
```

```python
#filtering by specific name
result=df[df["Name"]=="Svensson, Mr. Johan"]
result
```

```python
#counting occurences of variables

df["Sex"].value_counts()
```

```python
#count number of unique values

df.nunique()
```

```python
#filtering using 2 attributes

female=df["Sex"]=="female"
young=df["Age"]<50
female_young=df[female & young]
female_young
```

```python
#filtering using OR operator

old=df["Age"]>70
male=df["Sex"]=="male"
df[male | old]
```

```python
df["Sex"].value_counts()
```

```python
null=df["Age"].isnull()
df[null]
```

```python
#adding up all the nulls
df.isnull().sum()
```

```python
#dropping a column

df.drop(labels=["Pclass"], axis=1).head()
```

```python
#replacing NaN
df=df.replace("NaN", df["Age"].median())
df.replace("Masselmani, Mrs. Fatima", "Tanu")
```

```python
df.isnull().sum()
```

```python
#finding how many survived
df["Survived"].value_counts()/len(df)*100
```

```python
#chart survived vs died

%matplotlib inline
color=0.5
df["Survived"].value_counts().plot(kind="bar", title="chart")

```
