```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.ratemyprofessors.com/ShowRatings.jsp?tid=155191&showMyProfs=true"
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text,"html.parser")
soup
```

```python
#To find and count the Comments (tag) of the professor use this.

proftags = soup.findAll("span", {"class": "tag-box-choosetags" })
proftags
```

```python
#To find the ratings of the professor
proftags1 = soup.findAll("span", {"class": "rating-type" })
proftags1
```

```python

proftags2 = soup.findAll("span", {"class": "score poor"})
proftags2
```

```python
proftags2 = soup.findAll("span", {"class": "score average"})
proftags2
```

```python
proftags2 = soup.findAll("p", {"class": "commentsParagraph"})
proftags2
```

```python
#Finally to parse it as a text use this

for mytag in proftags:
  comments = mytag.get_text()
  print(comments)   
```
