```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")
```

```python
#adding content

dictionary={"Rank":[3, 2, 4, 5, 6],"Name":["Tanu", "Ashika", "Ashwin", "Mohit", "Sourabh"],"Age":[23, 24, 22, 18, 9]
            , "Weight":[69, 72, 65, 70, 40], "Occupation":["Student", "Working", "Working", "Student", "Student"]}
df=pd.DataFrame(dictionary)
df.head(5)
```

```python
#new dictionary
dictionary1 = {"Rank": [1], "Name": ["Tilo"], "Age": [80], "Weight": [80], "Occupation": ["Kitchen Queen"] }
df1=pd.DataFrame(dictionary1)
df1
```

```python
# Concatenating the two dataframes (df, df1)

df_row=pd.concat([df,df1])
df_row
```

```python
#remove the index
df_row.set_index('Rank', inplace=True)
df_row
```

```python
# To sort the data Rank in the ascending order
df_row.sort_values(by=['Rank'])
```

```python
#chart

df_row.plot()
plt.show()
```
